```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_of_grids = []
single_grids = []
for i in range(50):
    matrix = np.random.randint(2, size=(5, 5))
    matrix[2,2] = 0
    single_grids.append(matrix)
    list_of_grids.append(matrix)
    list_of_grids.append(np.rot90(matrix, 1))
    list_of_grids.append(np.rot90(matrix, 2))
    list_of_grids.append(np.rot90(matrix, 3))

```

```python
#make sure every grid in list_of_grids is unique
for i in range(200):
    for j in range(200):
        if i != j:
            if np.array_equal(list_of_grids[i], list_of_grids[j]):
                print("Duplicate found at index " + str(i) + " and " + str(j))
                print("Grids are: ")
                print(list_of_grids[i])
                print(list_of_grids[j])
```

```python
for i in range(len(single_grids)):
    #show the grid in black and white
    plt.imshow(single_grids[i], cmap='gray')
    #Add the labels on the sides 'a' on left 'b' on right ' 'c' on top 'd' on bottom
    plt.text(-1, 2, 'a', fontsize=12, rotation=90)
    plt.text(5, 2, 'b', fontsize=12, rotation=270)
    plt.text(2, 5, 'c', fontsize=12,rotation=180 )
    plt.text(2, -1, 'd', fontsize=12)
    plt.text(-1,-1, "Card {}".format(i + 1), fontsize=12)
    #remove the axis
    plt.axis('off')
    #make the orientaiton of the letters dependent on how you are holding the card 
    plt.savefig('NewCards/grid' + str(i) + '.png', bbox_inches='tight', pad_inches=0.5)
    
    plt.show()



    
```
